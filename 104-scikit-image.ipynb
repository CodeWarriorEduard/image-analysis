```python
from skimage import transform
from skimage.util import random_noise
from skimage.io import imread
import matplotlib.pyplot as plt

import numpy as np
```

```python
im = imread('./image_test/Hands-On-Image-Processing-with-Python-Second-Edition/Chapter01/images/hill.png')
tform = transform.SimilarityTransform(scale=0.9, rotation=np.pi/4, translation=(im.shape[0]/2, -100))
warped = transform.warp(im, tform)
plt.imshow(warped), plt.axis('off'), plt.show()
```

### Swirl transform

```python
im = imread('./image_test/Hands-On-Image-Processing-with-Python-Second-Edition/Chapter01/images/hill.png')
swirle = transform.swirl(im, rotation=0, strength=15, radius=200)
plt.imshow(swirle)
plt.axis('off')
plt.show()
```

### Random gaussian noise to images

```python
im = imread('./image_test/Hands-On-Image-Processing-with-Python-Second-Edition/Chapter01/images/hill.png')
sigmas = [0.1, 0.25, 0.5, 1]

for i in range(4):
    noise = random_noise(im, var=sigmas[i]**2)
    plt.subplot(2,2,i+1)
    plt.imshow(noise)
```

### Image manipulation with matplotlib


```python
from skimage.color import rgb2gray
im = rgb2gray(imread('./image_test/Hands-On-Image-Processing-with-Python-Second-Edition/Chapter01/images/me_now.jpg'))
plt.subplot(131)
plt.imshow(im, cmap='gray')
plt.subplot(132)
plt.contour(np.flipud(im), colors='k', levels=np.logspace(-15,15,100))
plt.subplot(133)
plt.contour(np.flipud(im), cmap='inferno')
```
