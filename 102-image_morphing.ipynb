```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from skimage import transform
```

```python
im1 = mpimg.imread('./image_test/Hands-On-Image-Processing-with-Python-Second-Edition/Chapter01/images/cycling.png')
im2 = mpimg.imread('./image_test/Hands-On-Image-Processing-with-Python-Second-Edition/Chapter01/images/girl.png')


im2 = transform.resize(im2, im1.shape, anti_aliasing=True)

i = 1

plt.figure(figsize=(18,15))

for alpha in np.linspace(0,1,20): ## Genera 20 valores entre 0 y 1
    plt.subplot(4,5, i) ## 4 filas 5 columnas i posicion
    plt.imshow((1-alpha)*im1 + alpha * im2) ## 
    plt.axis('off')
    i += 1
plt.subplots_adjust(wspace=0.05, hspace=0.05)
plt.show()
```
